```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from scipy.sparse import coo_matrix
```

```python
data = pd.read_csv('data/matrix.csv')
adj_matrix = data.iloc[:, 1:].values

matrix = coo_matrix(adj_matrix)
edges = np.r_[[matrix.row], [matrix.col]].T
G = nx.Graph()
G.add_nodes_from(range(adj_matrix.shape[0]))
G.add_edges_from(edges)
```

```python
first_names = data.Name.apply(lambda x: x.split(" ")[1])

fig, ax = plt.subplots(figsize=(50, 50))
nx.draw(G, labels=first_names, font_color='blue', node_color='orange')
plt.title("  ")
plt.margins(.2)
plt.show()
```

```python
names = data.Name
```

```python
sim = nx.betweenness_centrality(G)
print("  : ")
for k, v in sorted(sim.items(), key=lambda x: x[1], reverse=True)[:15]:
    print(names[k], round(v, 3))
```

```python
sim = nx.closeness_centrality(G)
print("  : ")
for k, v in sorted(sim.items(), key=lambda x: x[1], reverse=True)[:15]:
    print(names[k], round(v, 3))
```

```python
sim = nx.eigenvector_centrality(G)
print("   ")
for k, v in sorted(sim.items(), key=lambda x: x[1], reverse=True)[:15]:
    print(names[k], round(v, 3))
```
